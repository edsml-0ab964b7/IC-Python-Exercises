```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
img_black = np.zeros(shape=(800, 800, 3), dtype=np.int16)
plt.imshow(img_black)
```

```python
cv2.rectangle(img=img_black, pt1=(100,100), pt2=(400,300), color=(0,255,0), thickness=10)
plt.imshow(img_black)
```

```python
cv2.circle(img=img_black,center=(500,200),radius=50,color=(0,255,255),thickness=-1)
cv2.circle(img_black,(600,600), 100, (255, 0, 255) ,10)                     #Color is by default black\
plt.imshow(img_black)
```

```python
#Import only if not previously imported
import cv2
# Coordinates must be a tuple - (x,y)
cv2.line(img_black,(100,400),(600,600),(255,0,0),10)                   #Color is by default black
plt.imshow(img_black)
```

```python

font = cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.putText(img_black, "text", (200,600),font, 4, (255, 255, 255),5, cv2.LINE_AA)
plt.imshow(img_black)
```
